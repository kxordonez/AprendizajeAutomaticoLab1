## Data augmentation

```python
'''Los documentos de ImageDataGenerator sugieren que no se realiza ningún aumento de forma predeterminada'''
datagen = ImageDataGenerator(
        #rango de rotacion
        rotation_range=30,
        #rango especifico en el que se generaran con relacion al ancho
        width_shift_range=0.2,
        #rango especifico en el que se generaran con relacion a la altura
        height_shift_range=0.2,
        #escala de las imagenes
        rescale=1./255,
        shear_range=0.2,
        #acercamiento de las imagenes
        zoom_range=0.2,
        #direccion de las imagenes
        horizontal_flip=True,
        #sirve para llenar zonas enteras de bloques en 1 segundo, pero también puedes crear un bloque gigante de dinamita en 1 segundo
        fill_mode='nearest')
```

#### Con  mascara

```python
'''proporciona funciones para interactuar con el sistema operativo, es decir que va a tener navegabilidad entre directorios'''
os.chdir(r"C:/Users/lesly/Mineria/Unidad3/Proy_Mineria/Lab1/data_augmentation/con_mascara")
'''Iteracion ue permie obtener la lista de todos los archivos y directorios en el directorio especificado.'''
for path in os.listdir():
    #carga de cada una de las imagenes
    img = load_img(f"{path}")
    #creacion de un array Numpy con forma (3, 150, 150)
    x = img_to_array(img) 
    #Tamaño del array
    x = x.reshape((1,) + x.shape)
    #variable iterador
    i = 0
    #iterador para crear datos genericos
    for batch in datagen.flow(x, batch_size=1,
            save_to_dir=".",
            #nombre de las imagenes a las que va a recorrer
            save_prefix= 'con_mascara',
            #formato de las imagenes
            save_format='png'):
        #variable contador
        i += 1
        #crea siete imagenes por cada imagen que encuentre en la iteracion
        if i == 3:
            break 
```

## Filtro NLM para la eliminación de ruido de imágenes

```python
'''proporciona funciones para interactuar con el sistema operativo, es decir que va a tener navegabilidad entre directorios'''
os.chdir(r"C:/Users/lesly/Mineria/Unidad3/Proy_Mineria/Lab1")
'''Determinar la ruta en la que se encuentra la carpeta con las imagenes o datos.'''
images_path = "./data_augmentation"
'''Obtiene la lista de todos los archivos y directorios en el directorio especificado.'''
class_names = os.listdir(images_path)
''''Ordenar una lista en Python con el método sort'''
class_names.sort()
'''Imprimir el contenido de la variable class_names'''
print(class_names)
```

```python
'''Iterador encargado de recorrer las clases, es decir las carpetas creadas en el dataset '''
for class_name in class_names:
    cur_path = images_path + "/" + class_name
    cur_label = class_name
    #variable iterador
    i = 1
    #Tiene varias funciones que pueden ayudar a enumerar archivos en una carpeta especificada
    for file in glob.glob(cur_path + "/*"):
        #mensaje que se visualizara cuando se este realizando el procesamiento de cada una de las imagenes
        print("Processing Image - {} in {}".format(i, cur_label))
        # Leer la ruta de las imágenes
        image = cv2.imread(file)
        # eliminar el ruido de las imágenes
        dst = cv2.fastNlMeansDenoisingColored(image,None,10,10,7,21)
        b,g,r = cv2.split(dst)
        rgb_dst = cv2.merge([r,g,b])
        # guardar imagenes
        save_path = './filtra_ruido/'+ cur_label
        cv2.imwrite(os.path.join(save_path , cur_label + '_' + str(i) + '.png'),
                    cv2.cvtColor(rgb_dst,cv2.COLOR_RGB2BGR))
        

        #mostrar actualización de bucle
        i += 1
```

## Eliminación del ruido

```python
'''proporciona funciones para interactuar con el sistema operativo, es decir que va a tener navegabilidad entre directorios'''
os.chdir(r"C:/Users/lesly/Mineria/Unidad3/Proy_Mineria/Lab1")
'''Determinar la ruta en la que se encuentra la carpeta con las imagenes o datos.'''
images_path = "./filtra_ruido"
'''Obtiene la lista de todos los archivos y directorios en el directorio especificado.'''
class_names = os.listdir(images_path)
''''Ordenar una lista en Python con el método sort'''
class_names.sort()
'''Imprimir el contenido de la variable class_names'''
print(class_names)
```

```python
'''Iterador encargado de recorrer las clases, es decir las carpetas creadas en el dataset '''
for class_name in class_names:
    cur_path = images_path + "/" + class_name
    cur_label = class_name
    #variable iterador
    i = 1
    #Tiene varias funciones que pueden ayudar a enumerar archivos en una carpeta especificada
    for file in glob.glob(cur_path + "/*"):
        #mensaje que se visualizara cuando se este realizando el procesamiento de cada una de las imagenes
        print("Processing Image - {} in {}".format(i, cur_label))
        #lectura del directorio guardado en la variable file
        image = cv2.imread(file)
        #eliminacion del fondo
        output = remove(image)
        # Ruta en la que se van a guardar las imagenes sin fondo
        save_path = './eliminar_ruido/'+ cur_label
        cv2.imwrite(os.path.join(save_path , cur_label + '_' + str(i) + '.png'), output)
        # mostrar actualización de bucle
        i += 1
```
